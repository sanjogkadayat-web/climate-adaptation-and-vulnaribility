```python
import pycountry

YEARS = list(range(2010, 2024))  # 2010 through 2023 inclusive

# --- ND-GAIN composite score ---
gain_long = gain.melt(
    id_vars=["ISO3"],
    value_vars=[c for c in gain.columns if str(c).isdigit() and 2010 <= int(c) <= 2023],
    var_name="year",
    value_name="gain_score",
)
gain_long["year"] = gain_long["year"].astype(int)
gain_long = gain_long.rename(columns={"ISO3": "iso3"})
print("gain_long:", gain_long.shape)

# --- ND-GAIN vulnerability ---
vuln_long = vuln.melt(
    id_vars=["ISO3"],
    value_vars=[c for c in vuln.columns if str(c).isdigit() and 2010 <= int(c) <= 2023],
    var_name="year",
    value_name="vulnerability",
)
vuln_long["year"] = vuln_long["year"].astype(int)
vuln_long = vuln_long.rename(columns={"ISO3": "iso3"})
print("vuln_long:", vuln_long.shape)

# --- GDP per capita (current US$) ---
gdp_clean = gdp.loc[:, ~gdp.columns.str.contains("^Unnamed")]   # drop trailing empty col
gdp_long = gdp_clean.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="year",
    value_name="gdp_per_capita_usd",
)
gdp_long["year"] = pd.to_numeric(gdp_long["year"], errors="coerce")
gdp_long = gdp_long[gdp_long["year"].between(2010, 2023)].copy()
gdp_long["year"] = gdp_long["year"].astype(int)
gdp_long = gdp_long.rename(columns={"Country Code": "iso3"})[["iso3", "year", "gdp_per_capita_usd"]]
print("gdp_long:", gdp_long.shape)

# --- Government effectiveness ---
gov_clean = gov.dropna(subset=["Series Name"]).copy()
gov_clean.columns = [c.split(" ")[0] if "[YR" in c else c for c in gov_clean.columns]
gov_long = gov_clean.melt(
    id_vars=["Country Name", "Country Code", "Series Name", "Series Code"],
    var_name="year",
    value_name="gov_effectiveness",
)
gov_long["year"] = pd.to_numeric(gov_long["year"], errors="coerce")
gov_long = gov_long.dropna(subset=["year"]).copy()
gov_long["year"] = gov_long["year"].astype(int)
gov_long["gov_effectiveness"] = pd.to_numeric(gov_long["gov_effectiveness"], errors="coerce")  # ".." -> NaN
gov_long = gov_long.rename(columns={"Country Code": "iso3"})[["iso3", "year", "gov_effectiveness"]]
print("gov_long:", gov_long.shape)

# --- Income classification ---
inc_clean = inc.dropna(subset=["Unnamed: 0"]).copy()
inc_clean = inc_clean.rename(columns={"Unnamed: 0": "iso3", "Data for calendar year :": "country_name"})
inc_long = inc_clean.melt(
    id_vars=["iso3", "country_name"],
    var_name="year",
    value_name="income_group",
)
inc_long["year"] = pd.to_numeric(inc_long["year"], errors="coerce")
inc_long = inc_long.dropna(subset=["year"]).copy()
inc_long = inc_long[inc_long["year"].between(2010, 2023)].copy()
inc_long["year"] = inc_long["year"].astype(int)
inc_long["income_group"] = inc_long["income_group"].replace("..", pd.NA)
inc_long = inc_long[["iso3", "year", "income_group"]]
print("inc_long:", inc_long.shape)
```

```python
# --- OECD: combine Principal + Significant into a single aid total per country-year ---
oecd_long = (
    rio.groupby(["RECIPIENT", "TIME_PERIOD"], as_index=False)["OBS_VALUE"]
    .sum()
    .rename(columns={
        "RECIPIENT": "iso3",
        "TIME_PERIOD": "year",
        "OBS_VALUE": "adaptation_aid_usd_m",
    })
)

print("oecd_long:", oecd_long.shape)
print("years:", sorted(oecd_long["year"].unique()))
print("unique recipient countries:", oecd_long["iso3"].nunique())

print("\nTop 10 recipients (total adaptation aid 2010–2023, USD millions):")
print(
    oecd_long.groupby("iso3")["adaptation_aid_usd_m"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .round(1)
)

print("\nBottom of distribution:")
print(oecd_long["adaptation_aid_usd_m"].describe().round(2))
```

```python
def name_to_iso3(name):
    """Resolve country name to ISO3 alpha-3 code via pycountry. Returns None on failure."""
    if pd.isna(name):
        return None
    name = name.strip()
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        pass
    try:
        matches = pycountry.countries.search_fuzzy(name)
        if matches:
            return matches[0].alpha_3
    except LookupError:
        pass
    return None


# Manual fixes dictionary — add entries here for any names pycountry can't resolve
MANUAL_FIXES = {
    # "Some Country Name": "XYZ",
}


def resolve_list(df, label):
    df = df.copy()
    df["country_name"] = df["country_name"].str.strip()
    df["iso3"] = df["country_name"].map(name_to_iso3)

    # Apply manual overrides only if any are defined (avoids dtype issues with empty dict)
    if MANUAL_FIXES:
        df["iso3"] = df["iso3"].fillna(df["country_name"].map(MANUAL_FIXES))

    matched = df["iso3"].notna().sum()
    unmatched = df.loc[df["iso3"].isna(), "country_name"].tolist()
    print(f"{label}: {matched}/{len(df)} matched")
    if unmatched:
        print(f"  UNMATCHED → {unmatched}")
    return df
```

```python
lldc = resolve_list(lldc, "landlocked")
sids = resolve_list(sids, "sids")

# Build sets of ISO3 codes for use as binary flags in the merge
lldc_iso3 = set(lldc["iso3"].dropna())
sids_iso3 = set(sids["iso3"].dropna())
print(f"\nFinal: {len(lldc_iso3)} landlocked ISO3 codes, {len(sids_iso3)} SIDS ISO3 codes")
```

```python
MANUAL_FIXES = {
    "Lao People's Dem. Republic": "LAO",
    "St. Kitts and Nevis": "KNA",
    "St. Lucia": "LCA",
    "St. Vincent and the Grenadines": "VCT",
}
```

```python
# --- Merge all sources, anchored on OECD (recipients of adaptation aid) ---
panel = oecd_long.copy()
panel = panel.merge(gain_long,  on=["iso3", "year"], how="left")
panel = panel.merge(vuln_long,  on=["iso3", "year"], how="left")
panel = panel.merge(gdp_long,   on=["iso3", "year"], how="left")
panel = panel.merge(gov_long,   on=["iso3", "year"], how="left")
panel = panel.merge(inc_long,   on=["iso3", "year"], how="left")

# --- Binary flags from the LLDC and SIDS lists ---
panel["is_landlocked"] = panel["iso3"].isin(lldc_iso3).astype(int)
panel["is_sids"]       = panel["iso3"].isin(sids_iso3).astype(int)

# --- Add country names via pycountry (for readability in EDA / app) ---
def iso3_to_name(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except (AttributeError, KeyError):
        return None

panel["country"] = panel["iso3"].map(iso3_to_name)

# --- Reorder columns ---
panel = panel[[
    "iso3", "country", "year",
    "adaptation_aid_usd_m",
    "vulnerability", "gain_score",
    "gdp_per_capita_usd", "gov_effectiveness", "income_group",
    "is_landlocked", "is_sids",
]]

# --- Sanity checks ---
print("panel shape:", panel.shape)
print(f"unique countries: {panel['iso3'].nunique()}")
print(f"year range: {panel['year'].min()}–{panel['year'].max()}\n")

print("Missing values per column:")
print(panel.isna().sum())

print(f"\nUnique landlocked in panel: {panel.loc[panel['is_landlocked']==1, 'iso3'].nunique()}")
print(f"Unique SIDS in panel:        {panel.loc[panel['is_sids']==1, 'iso3'].nunique()}")

print("\nTop 5 recipients in 2023 with full merged data:")
print(
    panel.query("year == 2023")
    .nlargest(5, "adaptation_aid_usd_m")
    [["country", "adaptation_aid_usd_m", "vulnerability", "gdp_per_capita_usd", "gov_effectiveness", "income_group", "is_sids", "is_landlocked"]]
    .round(2)
    .to_string(index=False)
)

# --- Save the processed panel ---
PROCESSED.mkdir(parents=True, exist_ok=True)
panel.to_csv(PROCESSED / "panel.csv", index=False)
print(f"\nSaved → {PROCESSED / 'panel.csv'}")
```

```python
# 1. What are the 28 unmatched recipients?
print("Recipients with no country name (likely OECD regional codes, not countries):")
print(sorted(panel[panel["country"].isna()]["iso3"].unique()))

# 2. Aid distribution at a glance
print(f"\nAid (USD M) — mean ${panel['adaptation_aid_usd_m'].mean():.1f}, "
      f"median ${panel['adaptation_aid_usd_m'].median():.1f}, "
      f"max ${panel['adaptation_aid_usd_m'].max():.1f}")

# 3. THE central Week 1 question: does aid track vulnerability?
clean = panel.dropna(subset=["vulnerability", "adaptation_aid_usd_m"]).copy()
clean = clean[clean["adaptation_aid_usd_m"] > 0]
clean["log_aid"] = np.log(clean["adaptation_aid_usd_m"])

print(f"\nCorrelation (all years 2010–2023, n={len(clean)}):")
print(f"  vulnerability ↔ log(aid):  r = {clean['vulnerability'].corr(clean['log_aid']):.3f}")
print(f"  gain_score    ↔ log(aid):  r = {clean['gain_score'].corr(clean['log_aid']):.3f}")

# Pre/post Paris Agreement split
pre  = clean[clean["year"] <= 2015]
post = clean[clean["year"] >  2015]
print(f"\n  Pre-2015  (n={len(pre):>4}):  vuln↔aid  r = {pre['vulnerability'].corr(pre['log_aid']):.3f}")
print(f"  Post-2015 (n={len(post):>4}):  vuln↔aid  r = {post['vulnerability'].corr(post['log_aid']):.3f}")
```
